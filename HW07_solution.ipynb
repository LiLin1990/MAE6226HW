```python
import math
import numpy
from scipy import optimize
from matplotlib import pyplot
%matplotlib inline
```

## Problem 1##

```python
# create circle to map
ngtheta = 50
theta = numpy.radians(numpy.linspace(0,360,ngtheta))

# constants
c = 1
a = c/4
```

```python
# part a
T = 0.05
H = 0.00
r = (0.25+T/(3*math.sqrt(3)))                   # circle radius
zeta_c = -4*a/(3*math.sqrt(3))*T + 1j*2*a*H     # circle center
zeta = r*numpy.exp(1j*theta)                    # coordinates of circle surface
z = (zeta+zeta_c) + a**2/(zeta+zeta_c)          # coordinates of airfoil surface

# interpolate upper and lower surfaces
x, y = z.real, z.imag
ind = numpy.argmin(x)                 # find the leading edge
x_u = x[0:ind+1][::-1]
y_u = y[0:ind+1][::-1]
x_l = x[ind:]
y_l = y[ind:]

if x_l[0] == x_l[1]:
    x_l = x_l[1:]
    y_l = y_l[1:]
yy_u = numpy.interp(numpy.linspace(-c/2,c/2),x_u,y_u)
yy_l = numpy.interp(numpy.linspace(-c/2,c/2),x_l,y_l)

# numerically calculate thickness and camber ratios
T_num = numpy.amax(yy_u-yy_l)/c
H_num = numpy.amax((yy_u+yy_l)/2)/c
print(T_num)
print(H_num)

# plot the grid points
pyplot.figure(figsize=(8,4))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.scatter(x, y, color='#000000', s=10, marker='o', linewidth=0)
pyplot.axis('equal');
```

## Problems 2-3 ##

```python
# Define polar grid
ngr = 12;
ngtheta = 50

# airfoil properties
c = 1
T = 0.10
H = 0.02
a = c/4
r_s = (0.25+T/(3*math.sqrt(3)))
zeta_c = -4*a/(3*math.sqrt(3))*T + 1j*2*a*H

theta = numpy.radians(numpy.linspace(0,360,ngtheta));
r = numpy.linspace(r_s,2*r_s,ngr);

R, Theta = numpy.meshgrid(r, theta)
Zeta = R*numpy.exp(1j*Theta)

# flow parameters
Uinf = 1.0                                     # free stream velocity
al = 0.0*math.pi/180                           # angle of attack
kappa = 2*math.pi*Uinf*r_s**2                  # doublet strength for circle with radius r_s
Gam = 4*math.pi*r_s*Uinf*numpy.sin(al+2*H)     # vortex strength

# compute the complex potential on the mesh grid and complex velocity on the surface
W = Uinf*Zeta + kappa/(2*math.pi)/Zeta + 1j*Gam/(2*math.pi)*numpy.log(Zeta)
w = Uinf - kappa/(2*math.pi)/Zeta[:,0]**2 + 1j*Gam/(2*math.pi)/Zeta[:,0]
Zeta = Zeta*numpy.exp(1j*al)

# Joukowski transformation
Z = (Zeta+zeta_c) + a**2/(Zeta+zeta_c)
w = w/(1 - a**2/(Zeta[:,0]+zeta_c)**2)
cp = 1 - numpy.abs(w)**2/Uinf**2

# plot the streamlines and surface pressure coefficient
pyplot.figure(figsize=(16, 8))
pyplot.subplot(121)
pyplot.xlabel('$x$', fontsize=24)
pyplot.ylabel('$y$', fontsize=24)
pyplot.contour(Z.real, Z.imag, W.imag, 25, linewidths=2, linestyles='solid')
pyplot.plot(Z.real[:,0],Z.imag[:,0], color='#000000', linewidth=4);

pyplot.subplot(122)
pyplot.gca().invert_yaxis()
pyplot.xlabel('$x$', fontsize=24)
pyplot.ylabel('$c_p$', fontsize=24)
pyplot.plot(Z.real[:25,0],cp[:25],'b')
pyplot.plot(Z.real[25:,0],cp[25:],'r');
```
